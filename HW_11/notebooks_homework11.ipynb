```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x')
```

```python
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
k = f - g
```

```python
i = solve(k)
```

```python
[[x_elem.n().round(2),f.subs(x,x_elem).n().round(2)] for x_elem in i]
```

```python
plot(k)
```
